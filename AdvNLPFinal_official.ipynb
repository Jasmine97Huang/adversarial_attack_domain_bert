#Financial

```python
data = pd.read_csv("/content/data/FinancialPhraseBank-v1.0/Sentences_AllAgree.txt",\
                   encoding = "ISO-8859-1", sep="\@", header=None, names=["text", "label"])
```

```python
data.label.value_counts()
```

```python
data_sample = data.groupby("label", group_keys=False).apply(lambda df: df.sample(min(min(data.groupby("label").count()['text']), 50), random_state = 7)).reset_index(drop = True)
labels = {'neutral': 0, 'positive': 1, 'negative': 2}
#labels = {'positive':0,'negative':1,'neutral': 2}
data_sample['label'] = data_sample.label.apply(lambda x: labels[x])
```

```python
data_sample.label.value_counts()
```

```python
fin_data = []
for i, data in data_sample.iterrows():
    fin_data.append((data['text'], data['label']))

financial = textattack.datasets.Dataset(fin_data)
```

## yiyanghkust/finbert-tone
Pre-trained on financial communication text and fine-tuned on sentiment classifications task.

### Benchmark: 0.9

```python
finbert = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-tone',num_labels=3)
tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')

sentences = data_sample.text.tolist()

inputs = tokenizer(sentences, return_tensors="pt", padding=True)
outputs = finbert(**inputs)[0]
predicted = []

for idx, sent in enumerate(sentences):
    predicted.append(np.argmax(outputs.detach().numpy()[idx]))
```

```python
accuracy = sum(data_sample.label == predicted)/len(data_sample)
print("Benchmark accuracy is", accuracy)
```
